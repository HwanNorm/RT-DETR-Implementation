```python
import torch.nn as nn
import torchvision.transforms as T
from src.core import YAMLConfig
from PIL import Image
from google.colab.patches import cv2_imshow
```

```python
class Model(nn.Module):
    def __init__(self):
        super().__init__()

        cfg = YAMLConfig("/content/RT-DETR/rtdetrv2_pytorch/configs/rtdetrv2/rtdetrv2_r18vd_120e_coco.yml")
        checkpoint = torch.load("rtdetrv2_r18vd_120e_coco_rerun_48.1.pth", map_location='cpu')
        state = checkpoint['ema']['module'] if 'ema' in checkpoint else checkpoint['model']
        cfg.model.load_state_dict(state)
        self.model = cfg.model.deploy()
        self.postprocessor = cfg.postprocessor.deploy()

    def forward(self, images, orig_target_sizes):
        outputs = self.model(images)
        outputs = self.postprocessor(outputs, orig_target_sizes)
        return outputs


model = Model()
print("Model initialized successfully!")
```

```python
COCO_CLASSES = [
    'person', 'bicycle', 'car', 'motorcycle', 'airplane', 'bus',
    'train', 'truck', 'boat', 'traffic light', 'fire hydrant',
    'stop sign', 'parking meter', 'bench', 'bird', 'cat', 'dog',
    'horse', 'sheep', 'cow', 'elephant', 'bear', 'zebra', 'giraffe',
    'backpack', 'umbrella', 'handbag', 'tie', 'suitcase', 'frisbee',
    'skis', 'snowboard', 'sports ball', 'kite', 'baseball bat',
    'baseball glove', 'skateboard', 'surfboard', 'tennis racket',
    'bottle', 'wine glass', 'cup', 'fork', 'knife', 'spoon', 'bowl',
    'banana', 'apple', 'sandwich', 'orange', 'broccoli', 'carrot',
    'hot dog', 'pizza', 'donut', 'cake', 'chair', 'couch',
    'potted plant', 'bed', 'dining table', 'toilet', 'tv', 'laptop',
    'mouse', 'remote', 'keyboard', 'cell phone', 'microwave', 'oven',
    'toaster', 'sink', 'refrigerator', 'book', 'clock', 'vase',
    'scissors', 'teddy bear', 'hair drier', 'toothbrush'
]
```

```python
def process_frame(frame, model, device='cuda'):
    with torch.no_grad():
        rgb_frame = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        pil_image = Image.fromarray(rgb_frame)

        w, h = pil_image.size
        orig_size = torch.tensor([w, h])[None].to(device)

        transforms = T.Compose([
            T.Resize((640, 640)),
            T.ToTensor(),
        ])

        im_data = transforms(pil_image)[None].to(device)
        labels, boxes, scores = model(im_data, orig_size)

        for i in range(len(boxes[0])):
            if scores[0][i] > 0.5:
                box = boxes[0][i].detach().cpu().numpy()
                label_idx = labels[0][i].item()
                score = scores[0][i].item()

                class_name = COCO_CLASSES[int(label_idx)]

                cv2.rectangle(frame,
                            (int(box[0]), int(box[1])),
                            (int(box[2]), int(box[3])),
                            (0, 255, 0), 2)

                text = f"{class_name}: {score:.2f}"
                cv2.putText(frame, text,
                           (int(box[0]), int(box[1] - 5)),
                           cv2.FONT_HERSHEY_SIMPLEX, 0.5,
                           (0, 255, 0), 2)

        return frame
```

```python
from IPython.display import display, Javascript
from google.colab.output import eval_js
from base64 import b64decode, b64encode
import numpy as np

def js_to_image(js_reply):
    image_bytes = b64decode(js_reply.split(',')[1])
    jpg_as_np = np.frombuffer(image_bytes, dtype=np.uint8)
    img = cv2.imdecode(jpg_as_np, flags=1)
    return img

def run_webcam(model):
    js = Javascript('''
        var video;
        var div = null;
        var stream;
        var captureCanvas;
        var imgElement;
        var pendingResolve = null;
        var shutdown = false;

        function removeDom() {
           stream.getVideoTracks()[0].stop();
           video.remove();
           div.remove();
           video = null;
           div = null;
           stream = null;
           imgElement = null;
           captureCanvas = null;
        }

        function onAnimationFrame() {
          if (!shutdown) {
            window.requestAnimationFrame(onAnimationFrame);
          }
          if (pendingResolve) {
            var result = "";
            if (!shutdown) {
              captureCanvas.getContext('2d').drawImage(video, 0, 0, captureCanvas.width, captureCanvas.height);
              result = captureCanvas.toDataURL('image/jpeg', 0.8)
            }
            var lp = pendingResolve;
            pendingResolve = null;
            lp(result);
          }
        }

        async function createDom() {
            if (div !== null) {
                return stream;
            }

            // Get the device's screen width
            const screenWidth = window.innerWidth;

            // Calculate dimensions to maintain aspect ratio
            const containerWidth = Math.min(screenWidth - 40, 1280); // Max width with some padding
            const videoWidth = (containerWidth - 20) / 2; // Split width between two feeds
            const videoHeight = (videoWidth * 3) / 4; // Maintain 4:3 aspect ratio

            // Create container for both videos
            const container = document.createElement('div');
            container.style.display = 'flex';
            container.style.gap = '20px';
            container.style.width = containerWidth + 'px';
            container.style.margin = '0 auto';
            document.body.appendChild(container);

            // Raw webcam feed
            div = document.createElement('div');
            div.style.border = '2px solid black';
            div.style.padding = '3px';
            div.style.width = videoWidth + 'px';
            div.style.height = videoHeight + 'px';
            container.appendChild(div);

            // Processed feed
            const processedDiv = document.createElement('div');
            processedDiv.style.border = '2px solid blue';
            processedDiv.style.padding = '3px';
            processedDiv.style.width = videoWidth + 'px';
            processedDiv.style.height = videoHeight + 'px';
            container.appendChild(processedDiv);

            // Set up video element
            video = document.createElement('video');
            video.style.width = '100%';
            video.style.height = '100%';
            video.style.objectFit = 'cover';
            video.setAttribute('playsinline', '');
            video.onclick = () => { shutdown = true; };
            div.appendChild(video);

            // Set up processed image element
            imgElement = document.createElement('img');
            imgElement.style.width = '100%';
            imgElement.style.height = '100%';
            imgElement.style.objectFit = 'cover';
            processedDiv.appendChild(imgElement);

            stream = await navigator.mediaDevices.getUserMedia(
                {video: {
                    facingMode: "user",  // Changed from "environment" to "user" for front camera
                    width: { ideal: videoWidth },
                    height: { ideal: videoHeight }
                }});
            video.srcObject = stream;
            await video.play();

            captureCanvas = document.createElement('canvas');
            captureCanvas.width = videoWidth;
            captureCanvas.height = videoHeight;

            window.requestAnimationFrame(onAnimationFrame);
            return stream;
        }

        async function stream_frame() {
            if (shutdown) {
                removeDom();
                shutdown = false;
                return '';
            }

            stream = await createDom();

            var result = await new Promise(function(resolve) {
                pendingResolve = resolve;
            });

            shutdown = false;
            return result;
        }
    ''')
    display(js)

    model.to('cuda' if torch.cuda.is_available() else 'cpu')
    model.eval()

    while True:
        try:
            frame = eval_js('stream_frame()')
            if not frame:
                break

            frame = js_to_image(frame)
            processed_frame = process_frame(frame, model)

            _, encoded_img = cv2.imencode('.jpg', processed_frame)
            encoded_img = b64encode(encoded_img).decode('utf-8')
            eval_js(f"imgElement.src = 'data:image/jpeg;base64,{encoded_img}'")

        except KeyboardInterrupt:
            break

    eval_js('shutdown = true')
```

```python
run_webcam(model)
```
